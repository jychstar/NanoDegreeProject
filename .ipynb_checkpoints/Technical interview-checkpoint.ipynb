## Question 1

Given two strings `s` and `t`, determine whether some anagram of `t` is a substring of `s`. For example: if `s = "udacity"` and `t = "ad"`, then the function returns `True`. Your function definition should look like: `question1(s, t)` and return a boolean `True` or `False`.

````python
def question1(s,t):
  '''
  :type s: str
  :type t: str
  :rtype: bool
  '''
  length = len(s)
  sub_length = len(t)
  if sub_length > length or length == 0 or sub_length == 0:
    return False
  
  d ={}  # use a dictionary to count the letter frequency
  for letter in t:
    if letter in d:
      d[letter] += 1
    else:
      d[letter] = 1
  
  for i in range(length-sub_length+1):
    if s[i] in t and anagram(s[i:i+sub_length], d):
      return True
  return False

def anagram(sub,d):
  for letter in sub:
    if letter in d:
      d[letter] -= 1
      if d[letter] == 0:
        del d[letter]
    else:
      return False
  return True

# test cases
print question1('udacity','ad')
print question1('udacity','yti')  
print question1('success','scss')  
print question1('success','')  
````
